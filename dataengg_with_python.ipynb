```python
import pandas as pd
import random

names =["alice","bob",'ramu','somu','madav','faraan','chom']
dept = ['cs','ec','mech','civil']

data =[]
for i in range (1,20) :
  record = {
      "student id": i ,
      "name": random.choice(names),
      "age": random.choice([20,21,22,23,24,None]),
      "dept" :random.choice(dept + ['ece']),
      "marks":random.randint(50,100),
      "attendance":random.choice([random.randint(70,100),None])

  }
  data.append(record)

df = pd.DataFrame(data)

df.to_csv("students_random.csv",index=True)
print("random dirty dataset saved ")
print(df.head(10))
```

```python

import pandas as pd

df = pd.read_csv("/content/students_random.csv")
print(" Original Dirty Data")
print(df.head (10))
df = df.drop_duplicates()

df["age"] = df["age"].fillna(df["age"].mean())


df["attendance"] = df["attendance"].fillna (df["attendance"].median())

df["dept"] = df["dept"].replace({"ece": "ec"})

print("\n Cleaned Data")
print(df.head(10))
```

```python

import pandas as pd

df = pd.read_csv("/content/students_random.csv")
print(" Original Dirty Data")
print(df.head (10))
df = df.drop_duplicates()

df["age"] = df["age"].fillna(df["age"].mean())


df["attendance"] = df["attendance"].fillna (df["attendance"].median())

df["dept"] = df["dept"].replace({"ece": "ec"})

print("\n Cleaned Data")
print(df.head(10))

#4 automation
if "marks" in df.columns:
  low_marks = df[df["marks"]<60]
  if not low_marks.empty :
    print("\n warning")
    print(low_marks[["student id","name","marks"]])
  else:
      print("no failing students")

print("\n cleaned data")
print(df.head)

df.to_csv("students_cleaned.csv",index=False)
```
