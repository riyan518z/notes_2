```python
import requests
url = "https://api.spacexdata.com/v4/launches"

response = requests.get(url)
data1 = response.json()

launcher = pd.DataFrame(data1)
launcher.head()
```

```python
df1 = launcher[['name','date_utc','success','details','rocket','launchpad']]
df1
```

```python
df1.rename(columns={'name':'mission_name', 'rocket':'rocket_id'}, inplace=True)
df1
```

```python
import requests
url = "https://api.spacexdata.com/v4/rockets"

response = requests.get(url)
data2 = response.json()

rock = pd.DataFrame(data2)
rock.head()
```

```python
df2 = rock[['name','cost_per_launch','id','country','company']]
df2
```

```python
final = df1.merge(df2,left_on='rocket_id',right_on='id',how='left')
final
```
