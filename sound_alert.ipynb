```python
import pandas as pd
import time
from IPython.display import Audio , display
url="https://raw.githubusercontent.com/velicki/Weather_Data_Analysis_Project/main/Weather_Data.csv"
wdf = pd.read_csv(url)

wdf
```

```python
wdf.columns
```

```python
wdf.columns = wdf.columns.str.strip().str.replace(' ','_').str.replace('/','_')
wdf.columns
```

```python
wdf['Date_Time'] = pd.to_datetime(wdf['Date_Time'])
wdf.info()
```

```python
wdf.isna().sum()
```

```python
def alert_high_temp(row):
  if(row['Temp_C'] > 3):
    print(f'Alert !!! temp too high : {row['Temp_C']} celsius')
    display(Audio('/content/iPhone Notification Sound.mp3',autoplay=True))
    return True
  else:
    print(f"temp is normal : {row['Temp_C']} celsius")
    return False



def stream_with_alert(data,delay=1):
  for index,row in data.iterrows():
    print(f"{row['Date_Time']} - Temperature : {row['Temp_C']} Celsius")
    if alert_high_temp(row):
      print("help temp alert triggered . stopping streamprocessing")
      break
    time.sleep(delay)


stream_with_alert(wdf)
```
